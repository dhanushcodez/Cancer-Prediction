```python

import numpy as np 
import pandas as pd 
import warnings
warnings.filterwarnings("ignore")
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python

df = pd.read_csv('C:\\Users\\dhanush\\Desktop\\breast-cancer.csv')
df
```

```python
breast=df.drop(["id"],axis=1)
breast
```

```python

x=breast.drop(["diagnosis"],axis=1)
x=x.fillna(0) 
x
```

```python

target=breast["diagnosis"]
target
```

```python

data = (x - np.min(x, axis=0)) / (np.max(x, axis=0) - np.min(x, axis=0))
data
```

```python

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(data,target,test_size=0.3,random_state=1)
```

```python

from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,Y_train)
```

```python
accuracy=knn.score(X_test,Y_test)
print(accuracy)
```

```python

score_list=[]
for each in range(1,12):
    knn2=KNeighborsClassifier(n_neighbors=each)
    knn2.fit(X_train,Y_train)
    score_list.append(knn2.score(X_test,Y_test))
    print("{} knn score: {}".format(each,knn2.score(X_test,Y_test)))
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,12),score_list,color='r')
plt.show()
```

```python

from sklearn.svm import SVC
from sklearn.metrics import accuracy_score 
svm=SVC(random_state=1)
svm.fit(X_train,Y_train)
Y_pred = svm.predict(X_test)
my_accuracy = accuracy_score(Y_test, Y_pred, normalize=True)
print(my_accuracy)
```

```python

Results=[]
for k in [1,2]:
    for n in range(1,3):
        for c in [0.5,1,2]:
            if k == 1:
                svclassifier = SVC(kernel="poly",degree=n,C=c)
                svclassifier.fit(X_train, Y_train)
                
            else:
                svclassifier = SVC(kernel="rbf",gamma=n,C=c)
                svclassifier.fit(X_train, Y_train)
                
            Y_pred = svclassifier.predict(X_test)
            my_accuracy = accuracy_score(Y_test, Y_pred, normalize=True) 
            
            Results.append((k,n,c,my_accuracy))
            print("accuracy:", Results)
            print()  
        print()  

print()  
```

```python

Results_sorted = sorted(Results, key=lambda x: x[3], reverse=True)

top_5 = Results_sorted[:5]
print("the top 5 models",top_5)

plt.figure(figsize=(8,6))
for i, result in enumerate(top_5):
    plt.plot(i, result[3], 'ro', markersize=10)
    plt.text(i+0.1, result[3]+0.005, f"k={result[0]}, n={result[1]}, c={result[2]}")
plt.ylim([0.7, 1.0])
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Top 5 models')
plt.show()
```

```python

from sklearn.naive_bayes import GaussianNB
gaussian=GaussianNB()
gaussian.fit(X_train,Y_train)
```

```python
Y_pred = gaussian.predict(X_test)
my_accuracy = accuracy_score(Y_test, Y_pred, normalize=True)
print(my_accuracy)
```

```python

from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,Y_train)
```

```python
Y_pred = dt.predict(X_test)
my_accuracy = accuracy_score(Y_test, Y_pred, normalize=True)
print(my_accuracy)
```

```python

best_accuracy = 0.0
best_params = {}


for max_depth in [2, 4, 6]:
    for min_samples_split in [2, 4, 6]:
        for criterion in ["gini", "entropy"]:
            
            dt2= DecisionTreeClassifier(max_depth=max_depth, min_samples_split=min_samples_split, criterion=criterion)
            dt2.fit(X_train, Y_train)
            
            Y_pred = dt2.predict(X_test)
            
            current_accuracy = accuracy_score(Y_test, Y_pred, normalize=True)
            
            if current_accuracy > best_accuracy:
                best_accuracy = current_accuracy
                best_params = {
                    'max_depth': max_depth,
                    'min_samples_split': min_samples_split,
                    'criterion': criterion
                }

print("Best Accuracy:", best_accuracy)
print("Best Parameters:", best_params)
```

```python

from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train,Y_train)
```

```python
Y_pred=rf.predict(X_test)
my_accuracy=accuracy_score(Y_test,Y_pred, normalize=True)
print(my_accuracy)
```

```python
best_accuracy = 0.0
best_params = {}

for n_estimators in [10, 30, 50]:
    for max_depth in [2, 4, 6]:
        for max_features in ['sqrt', 'log2']:
            for min_samples_split in [2, 4, 6]:
                
                rf2 = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth,
                                            max_features=max_features, min_samples_split=min_samples_split)
                rf2.fit(X_train, Y_train)
                
              
                Y_pred = rf2.predict(X_test)
                
                current_accuracy = accuracy_score(Y_test, Y_pred, normalize=True)
                
                
                if current_accuracy > best_accuracy:
                    best_accuracy = current_accuracy
                    best_params = {
                        'n_estimators': n_estimators,
                        'max_depth': max_depth,
                        'max_features': max_features,
                        'min_samples_split': min_samples_split
                    }

print("Best Accuracy:", best_accuracy)
print("Best Parameters:", best_params)
```
